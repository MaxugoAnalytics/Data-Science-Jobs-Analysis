### Data Analysis using different questions from client

```python
# Who earns most salary based on experience level
Average_Salary_by_experience_level = justice.groupby(['experience_level'])['salary_in_usd'].mean().sort_values(ascending=False).reset_index()
Average_Salary_by_experience_level

```

```python
# rename 'salary_in_usd' to 'average_salary_by_EL'
Average_Salary_by_experience_level.rename(columns = {'salary_in_usd':'average_salary_by_EL'},inplace=True)
Average_Salary_by_experience_level
```

```python
plt.pie(Average_Salary_by_experience_level['average_salary_by_EL'], 
        labels= Average_Salary_by_experience_level['experience_level'],
        autopct='%1.1f%%')
plt.title('PERCENTAGE OF AVERAGE SALARY BY EXPERIENCE LEVEL')
plt.show()
```

```python
Average_Salary_by_experience_level.plot(kind='bar',x='experience_level',y='average_salary_by_EL',width=0.75,color='green')   
```

```python
# What "job_title" earns the most salary
Average_salary_by_job_title=justice.groupby(['job_title'])['salary_in_usd'].mean().sort_values(ascending=False)[:10].reset_index()
Average_salary_by_job_title
```

```python
Average_salary_by_job_title.plot(kind='bar',x='job_title',y='salary_in_usd',width=0.75,color='green')
```

```python
# What is salary based on company size
Average_salary_by_company= justice.groupby(['company_size'])['salary_in_usd'].mean().sort_values(ascending=False).reset_index()
Average_salary_by_company
```

```python
# What company location pay the most salary
Average_salary_by_location= justice.groupby(['company_location',
'job_title'])['salary_in_usd'].mean().sort_values(ascending=False)[:10].reset_index()
Average_salary_by_location
```

```python
# What city/country employees reside and earn most salaries and what employment type 
location_experience=justice.groupby(['employee_residence',
'job_title','employment_type'])['salary_in_usd'].mean().sort_values(ascending=False)[:10].reset_index()
location_experience
```

```python
# What is the most common 'job type' accross all countries/city

Most_job = justice['job_title'].value_counts()[:5].reset_index()
Most_job.rename(columns = {'job_title':'total_job'},inplace=True)
Most_job
```

```python
Most_job.rename(columns = {'index':'job_title'},inplace=True)
Most_job
```

```python
plt.pie(Most_job['total_job'], 
        labels=Most_job['job_title'],
        autopct='%1.1f%%')
plt.title('Percentage of most jobs accross city/country')
plt.show()
```

```python
# Which country has most Jobs
Most_job_country = justice['company_location'].value_counts()[:10].reset_index()
Most_job_country
```

```python
# What are the most job title in the US
US_job = justice.loc[justice['company_location']=='US','job_title'].value_counts()[:10]
US_job
```

```python
US_job.sum()
```

```python
# Who earns(top ten)  the most salary in the US
US_job_salary = justice.loc[(justice['employee_residence']=='US')&
                            (justice['salary_in_usd'].notna()),
                            ['job_title',
            'salary_in_usd']].groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False)[:10].reset_index()
US_job_salary
```
